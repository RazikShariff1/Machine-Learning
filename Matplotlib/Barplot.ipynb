## Bar plot

```python
import matplotlib.pyplot as pl
```

```python
x=["logistics regression","random forest","svm","Neural Network"]
y=[85,92,88,95]
pl.xlabel("Machine Learning Models")
pl.ylabel("Accuracy (%)")
pl.title("Model Performance Comparison")
pl.bar(x,y,color='skyblue',edgecolor='black',alpha=0.8, linewidth=2, 
        width=0.5, 
        hatch='o', #can be anything 
        label="Accuracy" # position the lable
      )
pl.show()
```

```python
pl.barh(x, y, color='lightcoral', edgecolor='black', height=0.5)
```

```python
import matplotlib.pyplot as plt  

models = ["Logistic Regression", "Random Forest", "SVM", "Neural Network"]
accuracies = [85, 92, 88, 95]

# Assigning different colors to each bar
colors = ['red', 'green', 'blue', 'purple']

plt.bar(models, accuracies, color=colors,alpha=.5)

plt.xlabel("Models")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.title("Bar Color Customization")

plt.show()
```


# 🔥 **All `plt.bar()` Parameters & Their Possible Inputs**  

| Parameter | Type | Possible Inputs | Description |
|-----------|------|----------------|-------------|
| `x` | List/Array | `["A", "B", "C"]`, `np.array([1,2,3])` | X-axis categories or positions |
| `height` | List/Array | `[10, 20, 30]`, `np.array([15, 25, 35])` | Heights of bars |
| `color` | String / List | `'blue'`, `['red', 'green']`, `'#ff5733'` | Fill color of bars |
| `edgecolor` | String / List | `'black'`, `['red', 'blue']` | Border color of bars |
| `linewidth` | Float / Int | `1.5`, `2`, `0.5` | Border thickness |
| `alpha` | Float (0-1) | `0.1` (10% opacity), `0.5`, `1` (no transparency) | Transparency of bars |
| `width` | Float / List | `0.5`, `1`, `[0.2, 0.4, 0.6]` | Width of bars |
| `align` | String | `'center'`, `'edge'` | Bar alignment on x-axis |
| `hatch` | String | `'/'`, `'\\'`, `'*'`, `'x'`, `'-'` | Patterns inside bars |
| `bottom` | List/Array | `[0, 5, 10]`, `np.zeros(3)` | Y-start position (for stacked bars) |
| `label` | String | `'Dataset A'`, `'Accuracy'` | Legend label for bars |
