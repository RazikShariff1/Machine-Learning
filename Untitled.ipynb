```python
import matplotlib.pyplot as pl
pl.bar([2,4,6,8],[1,2,2,1],color=['r','b','g','y'],width=0.4,align='edge')
pl.show()
```

```python
# plt.plot(x, y)  # Plotting x vs y
plt.xlabel("X Values",fontsize=12)  # X-axis label
plt.ylabel("Y Values",fontsize=22)  # Y-axis label
plt.title("Basic Line Plot")  # Title
plt.show()  # Display the plot
```

```python
import matplotlib.pyplot as plt  

# Data for plotting
x = [1, 2, 3, 4, 5]  # X-axis values
y = [10, 20, 25, 30, 50]  # Y-axis values

# Create a simple line plot
plt.plot(x, y, label="Growth", color='blue', marker='o', linestyle='--')

# Add labels and title
plt.xlabel("Time")  
plt.ylabel("Value")  
plt.title("Basic Line Plot")  
plt.legend()  # Show legend

# Display the plot
plt.show()
```
